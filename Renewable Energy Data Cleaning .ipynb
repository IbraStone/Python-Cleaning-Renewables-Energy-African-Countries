#  Cleaning dataset Renewable Energy  in Africa from US Energy Information Administration

```python
# Load dataset and display dataset

Renewable_energy_AfricaPy = pd.read_csv('C:/Users/ibrah/OneDrive/Desktop/Data Analytics Certificate Program/Renewable_energy_AfricaPy.csv')
print(Renewable_energy_AfricaPy)
```

 #### Handling Missing data
 

```python
# Check for missing values in the entire dataset
missing_values = Renewable_energy_AfricaPy.isnull().sum()

# If you want to see the missing values as a percentage of the total for each column:
missing_percent = (missing_values / len(Renewable_energy_AfricaPy)) * 100

# Create a DataFrame to display the missing values and percentages
missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage': missing_percent})

# Print or display the missing data
print(missing_data)
```

```python
# Impute missing values in numerical columns with the mean
numerical_columns = Renewable_energy_AfricaPy.select_dtypes(include=['number'])
Renewable_energy_AfricaPy[numerical_columns.columns] = numerical_columns.fillna(numerical_columns.mean())

# Impute missing values in categorical columns with the mode
categorical_columns = Renewable_energy_AfricaPy.select_dtypes(include=['object'])
Renewable_energy_AfricaPy[categorical_columns.columns] = categorical_columns.fillna(categorical_columns.mode().iloc[0])

# Print the updated DataFrame
print(Renewable_energy_AfricaPy)
```

 #### Remove duplicate 

```python
# Use the .drop_duplicates() method to remove duplicate rows
Renewable_energy_AfricaPy = Renewable_energy_AfricaPy.drop_duplicates(keep='last')
# Print the updated DataFrame
print(Renewable_energy_AfricaPy)
```

 #### Drop columns

```python
print(Renewable_energy_AfricaPy.columns)
```

```python
# List the names of the columns you want to drop
columns_to_drop = ['1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', 
                   '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001',
                   '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009']

# Use the .drop() method to remove the specified columns
Renewable_energy_AfricaPy.drop(columns=columns_to_drop, inplace=True)

# Print the updated DataFrame
print(Renewable_energy_AfricaPy)
```

```python
# Replace 'df' with your actual DataFrame name
grouped_data = Renewable_energy_AfricaPy.groupby('Country').sum(numeric_only=True)

# Reset the index to have 'Country' as a regular column
grouped_data = grouped_data.reset_index()

# Print the resulting DataFrame
print(grouped_data)
```
